```python
train = pd.read_csv(os.path.join("train.csv"))
test = pd.read_csv(os.path.join("test.csv"))
sub = pd.read_csv(os.path.join("sample_submit.csv"),header=None)


print(train.shape)
display(train.head())
print(test.shape)
display(test.head())
sub.columns = ["id","attendance"]
display(sub.head())
```

```python
stadium_df = pd.read_csv(os.path.join("venue_information.csv"))
holiday_df = pd.read_csv(os.path.join("holidays_in_japan.csv"))
match_df = pd.read_csv(os.path.join("match_reports.csv"))
display(stadium_df.head())
display(holiday_df.head())
display(match_df.head())
```

```python
matchs = match_df[["id","home_team_score","away_team_score"]]
match_info_df = pd.concat([train,test],axis=0)
match_info_df = match_info_df[["id","match_date","home_team","away_team"]]
match_info_df = match_info_df.merge(matchs,how="left",on="id")
match_info_df['match_date'] = pd.to_datetime(match_info_df["match_date"])
match_info_df["diff_score"] = match_info_df["home_team_score"] - match_info_df["away_team_score"]

match_info_df["get_home_team_point"] = 1
match_info_df.loc[match_info_df["diff_score"]>0, "get_home_team_point"] = 3 
match_info_df.loc[match_info_df["diff_score"]<0, "get_home_team_point"] = 0

match_info_df["get_away_team_point"] = 1
match_info_df.loc[match_info_df["get_home_team_point"]==3, "get_away_team_point"] = 0
match_info_df.loc[match_info_df["get_home_team_point"]==0, "get_away_team_point"] = 3

match_info_df.drop("diff_score",axis=1, inplace=True)

display(match_info_df.head())
```

```python
#ホームチームのホームゲーム直近3試合の勝ち点、得点、失点
for team in list(match_info_df.home_team.unique()):
    home_info_df = match_info_df.loc[match_info_df["home_team"]==team]
    home_info_df = home_info_df.sort_values("match_date")
    home_info_df["Season"] = home_info_df.match_date.dt.year
    for season in list(home_info_df.Season.unique()):
        home_season_df = home_info_df.loc[home_info_df.Season==season]
        home_season_df["get_home_team_point"] = home_season_df["get_home_team_point"].shift(1).fillna(0)
        home_season_df["home_team_score"] = home_season_df["home_team_score"].shift(1).fillna(0)
        home_season_df["away_team_score"] = home_season_df["away_team_score"].shift(1).fillna(0)
        home_season_df["Curren3HomeMatch_point_sum"] = home_season_df["get_home_team_point"].rolling(3,min_periods=1).sum()
        #home_season_df["Curren3HomeMatch_point_mean"] = home_season_df["get_home_team_point"].rolling(3,min_periods=1).mean()
        home_season_df["Curren3HomeMatch_point_std"] = home_season_df["get_home_team_point"].rolling(3,min_periods=1).std().fillna(0)
        home_season_df["Curren3HomeMatch_get_score_sum"] = home_season_df["home_team_score"].rolling(3,min_periods=1).sum()
        #home_season_df["Curren3HomeMatch_get_score_mean"] = home_season_df["home_team_score"].rolling(3,min_periods=1).mean()
        home_season_df["Curren3HomeMatch_get_score_std"] = home_season_df["home_team_score"].rolling(3,min_periods=1).std().fillna(0)
        home_season_df["Curren3HomeMatch_lost_score_sum"] = home_season_df["away_team_score"].rolling(3,min_periods=1).sum()
        #home_season_df["Curren3HomeMatch_lost_score_mean"] = home_season_df["away_team_score"].rolling(3,min_periods=1).mean()
        home_season_df["Curren3HomeMatch_lost_score_std"] = home_season_df["away_team_score"].rolling(3,min_periods=1).std().fillna(0)
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_point_sum"] = home_season_df["Curren3HomeMatch_point_sum"]
        #home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_point_mean"] = home_season_df["Curren3HomeMatch_point_mean"]
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_point_std"] = home_season_df["Curren3HomeMatch_point_std"]
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_get_score_sum"] = home_season_df["Curren3HomeMatch_get_score_sum"]
        #home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_get_score_mean"] = home_season_df["Curren3HomeMatch_get_score_mean"]
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_get_score_std"] = home_season_df["Curren3HomeMatch_get_score_sum"]
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_lost_score_sum"] = home_season_df["Curren3HomeMatch_lost_score_sum"]
        #home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_lost_score_mean"] = home_season_df["Curren3HomeMatch_lost_score_mean"]
        home_info_df.loc[home_info_df.Season==season,"Curren3HomeMatch_lost_score_std"] = home_season_df["Curren3HomeMatch_lost_score_sum"]
    del home_season_df

    home_info_df["Season_home_point"] = home_info_df.groupby("Season")["get_home_team_point"].cumsum()
    home_info_df["Season_home_point"] = home_info_df.groupby("Season")["Season_home_point"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない
    home_info_df["Season_home_get_score"] = home_info_df.groupby("Season")["home_team_score"].cumsum()
    home_info_df["Season_home_get_score"] = home_info_df.groupby("Season")["Season_home_get_score"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない
    home_info_df["Season_home_lost_score"] = home_info_df.groupby("Season")["away_team_score"].cumsum()
    home_info_df["Season_home_lost_score"] = home_info_df.groupby("Season")["Season_home_lost_score"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない

    home_info_df["before_home_match_point"] = home_info_df.groupby("Season")["get_home_team_point"].shift(1).fillna(-1) #前のホーム試合の勝ち点
    home_info_df["before_home_match_get_score"] = home_info_df.groupby("Season")["home_team_score"].shift(1).fillna(-1) #前のホーム試合の得点
    home_info_df["before_home_match_lost_score"] = home_info_df.groupby("Season")["away_team_score"].shift(1).fillna(-1) #前のホーム試合の失点

    home_info_df["before_match_date"] = home_info_df["match_date"].shift(1)
    home_info_df["diff_match_date"] = (home_info_df["match_date"] - home_info_df["before_match_date"]).dt.days
    home_info_df["diff_match_date"] = home_info_df["diff_match_date"].fillna(0)

    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_point"] = home_info_df["Season_home_point"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_get_score"] = home_info_df["Season_home_get_score"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_lost_score"] = home_info_df["Season_home_lost_score"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_diff_date"] = home_info_df["diff_match_date"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_point_sum"] = home_info_df["Curren3HomeMatch_point_sum"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_point_mean"] = home_info_df["Curren3HomeMatch_point_mean"]   
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_point_std"] = home_info_df["Curren3HomeMatch_point_std"]   
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_get_score_sum"] = home_info_df["Curren3HomeMatch_get_score_sum"]   
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_get_score_mean"] = home_info_df["Curren3HomeMatch_get_score_mean"]   
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_get_score_std"] = home_info_df["Curren3HomeMatch_get_score_std"]   
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_lost_score_sum"] = home_info_df["Curren3HomeMatch_lost_score_sum"]   
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_lost_score_mean"] = home_info_df["Curren3HomeMatch_lost_score_mean"]   
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_HomeGame_Curren3HomeMatch_lost_score_std"] = home_info_df["Curren3HomeMatch_lost_score_std"]
    del home_info_df
match_info_df   
```

```python
#ホームチームとアウェイチームの直近4試合(だいたい1ヶ月分)の獲得勝ち点数の標準偏差の差
for team in list(match_info_df.home_team.unique()):
    home_df = match_info_df.loc[match_info_df["home_team"]==team]
    home_df["my_team_name"] = team
    home_df["my_team_score"] = home_df["home_team_score"]
    home_df["opponent_score"] = home_df["away_team_score"]
    home_df["get_my_team_point"] = home_df["get_home_team_point"]
    away_df = match_info_df.loc[match_info_df["away_team"]==team]
    away_df["my_team_name"] = team
    away_df["my_team_score"] = away_df["away_team_score"]
    away_df["opponent_score"] = away_df["home_team_score"]
    away_df["get_my_team_point"] = away_df["get_away_team_point"]
    team_df = pd.concat([home_df,away_df],axis=0)
    team_df = team_df.sort_values("match_date")
    team_df["Season"] = team_df.match_date.dt.year
    for season in list(team_df.Season.unique()):

        season_df = team_df.loc[team_df.Season==season]
        season_tmp_df = season_df.sort_values("match_date").reset_index(drop=True)
        season_df["Season_num_match"] = season_tmp_df.index + 1  #シーズン試合数
        season_df["get_my_team_point"] = season_df["get_my_team_point"].shift(1).fillna(0)
        season_df["my_team_score"] = season_df["my_team_score"].shift(1).fillna(0)
        season_df["opponent_score"] = season_df["opponent_score"].shift(1).fillna(0)
        season_df["Curren4match_point_sum"] = season_df["get_my_team_point"].rolling(4,min_periods=1).sum()
        season_df["Curren4match_point_mean"] = season_df["get_my_team_point"].rolling(4,min_periods=1).mean()
        season_df["Curren4match_point_std"] = season_df["get_my_team_point"].rolling(4,min_periods=1).std().fillna(0)
        season_df["Curren4match_get_score_sum"] = season_df["my_team_score"].rolling(4,min_periods=1).sum()
        season_df["Curren4match_get_score_mean"] = season_df["my_team_score"].rolling(4,min_periods=1).mean()
        season_df["Curren4match_get_score_std"] = season_df["my_team_score"].rolling(4,min_periods=1).std().fillna(0)
        season_df["Curren4match_lost_score_sum"] = season_df["opponent_score"].rolling(4,min_periods=1).sum()
        season_df["Curren4match_lost_score_mean"] = season_df["opponent_score"].rolling(4,min_periods=1).mean()
        season_df["Curren4match_lost_score_std"] = season_df["opponent_score"].rolling(4,min_periods=1).std().fillna(0)
        team_df.loc[team_df.Season==season,"Season_num_match"] = season_df["Season_num_match"]
        team_df.loc[team_df.Season==season,"Curren4match_point_sum"] = season_df["Curren4match_point_sum"]
        #team_df.loc[team_df.Season==season,"Curren4match_point_mean"] = season_df["Curren4match_point_mean"]
        team_df.loc[team_df.Season==season,"Curren4match_point_std"] = season_df["Curren4match_point_std"]
        team_df.loc[team_df.Season==season,"Curren4match_get_score_sum"] = season_df["Curren4match_get_score_sum"]
        #team_df.loc[team_df.Season==season,"Curren4match_get_score_mean"] = season_df["Curren4match_get_score_mean"]
        team_df.loc[team_df.Season==season,"Curren4match_get_score_std"] = season_df["Curren4match_get_score_sum"]
        team_df.loc[team_df.Season==season,"Curren4match_lost_score_sum"] = season_df["Curren4match_lost_score_sum"]
        #team_df.loc[team_df.Season==season,"Curren4match_lost_score_mean"] = season_df["Curren4match_lost_score_mean"]
        team_df.loc[team_df.Season==season,"Curren4match_lost_score_std"] = season_df["Curren4match_lost_score_sum"]
        del season_df

    team_df["Season_point"] = team_df.groupby("Season")["get_my_team_point"].cumsum()
    team_df["Season_point"] = team_df.groupby("Season")["Season_point"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない
    team_df["Season_get_score"] = team_df.groupby("Season")["my_team_score"].cumsum()
    team_df["Season_get_score"] = team_df.groupby("Season")["Season_get_score"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない
    team_df["Season_lost_score"] = team_df.groupby("Season")["opponent_score"].cumsum()
    team_df["Season_lost_score"] = team_df.groupby("Season")["Season_lost_score"].shift(1).fillna(0)  #試合前は勝ち点が確定していないため前の試合のまでの結果しか使えない

    team_df["before_match_point"] = team_df.groupby("Season")["get_my_team_point"].shift(1).fillna(-1) #前の試合の勝ち点
    team_df["before_match_get_score"] = team_df.groupby("Season")["my_team_score"].shift(1).fillna(-1) #前の試合の得点
    team_df["before_match_lost_score"] = team_df.groupby("Season")["opponent_score"].shift(1).fillna(-1) #前の試合の失点
  
    # カウントエンコーディングがだめならだめ
    # 全てのデータを使って近年のJ1での成績から変数を作成
    #team_df["all_point"] = team_df["get_my_team_point"].sum()  #ポイントが高いチームほど毎年勝っている(近年？強いチーム)
    #team_df["all_get_score_sum"] = team_df["my_team_score"].sum()  #得点が高いチームほどチームカラー的に点をたくさん取るチーム 
    #team_df["all_get_score_mean"] = team_df["my_team_score"].mean()
    #team_df["all_point/count_season"] = team_df["all_point"] / len(team_df.Season.unique())
    #team_df["all_score/count_season"] = team_df["all_get_score_sum"] / len(team_df.Season.unique())
    #team_df["Count_J1_Season"] = len(team_df.Season.unique())
    
    #ホームチームの直近行われたホームゲームの試合日と今回のホームゲームの日数差
    team_df["before_match_date"] = team_df["match_date"].shift(1)
    team_df["diff_match_date"] = (team_df["match_date"] - team_df["before_match_date"]).dt.days
    team_df["diff_match_date"] = team_df["diff_match_date"].fillna(0)

    team_df = team_df[["id","home_team","away_team","Season_num_match","Season_point","Season_get_score","Season_lost_score",
                     "Curren4match_point_sum","Curren4match_point_std",
                     "Curren4match_get_score_sum","Curren4match_get_score_std",
                     "Curren4match_lost_score_sum","Curren4match_lost_score_std",
                     "before_match_point","before_match_get_score","before_match_lost_score",
                     #"all_point","all_get_score_sum","all_point/count_season","all_score/count_season","Count_J1_Season",
                       "diff_match_date"]]
    home_df2 = team_df.loc[team_df["home_team"]==team].rename(columns={"Season_num_match":"home_team_season_num_match","Season_point":"home_team_season_point","Season_get_score":"home_team_season_get_score","Season_lost_score":"home_team_season_lost_score",
                                                                     "Curren4match_point_sum":"home_team_Curren4match_point_sum","Curren4match_point_std":"home_team_Curren4match_point_std",
                                                                     "Curren4match_get_score_sum":"home_team_Curren4match_get_score_sum","Curren4match_get_score_std":"home_team_Curren4match_get_score_std",
                                                                     "Curren4match_lost_score_sum":"home_team_Curren4match_lost_score_sum","Curren4match_lost_score_std":"home_team_Curren4match_lost_score_std",
                                                                     "before_match_point":"home_team_before_match_point","before_match_get_score":"home_team_before_match_get_score","before_match_lost_score":"home_team_before_match_lost_score",
                                                                     #"all_point":"home_team_all_point","all_get_score_sum":"home_team_all_get_score_sum",
                                                                     #"all_point/count_season":"home_team_all_point/count_season","all_score/count_season":"home_team_all_score/count_season","Count_J1_Season":"home_team_Count_J1_Season",
                                                                     "diff_match_date":"home_team_diff_match_date"
                                                                     })
    away_df2 = team_df.loc[team_df["away_team"]==team].rename(columns={"Season_num_match":"away_team_season_num_match","Season_point":"away_team_season_point","Season_get_score":"away_team_season_get_score","Season_lost_score":"away_team_season_lost_score",
                                                                     "Curren4match_point_sum":"away_team_Curren4match_point_sum","Curren4match_point_std":"away_team_Curren4match_point_std",
                                                                     "Curren4match_get_score_sum":"away_team_Curren4match_get_score_sum","Curren4match_get_score_std":"away_team_Curren4match_get_score_std",
                                                                     "Curren4match_lost_score_sum":"away_team_Curren4match_lost_score_sum","Curren4match_lost_score_std":"away_team_Curren4match_lost_score_std",
                                                                     "before_match_point":"away_team_before_match_point","before_match_get_score":"away_team_before_match_get_score","before_match_lost_score":"away_team_before_match_lost_score",
                                                                     #"all_point":"away_team_all_point","all_get_score_sum":"away_team_all_get_score_sum",
                                                                     #"all_point/count_season":"away_team_all_point/count_season","all_score/count_season":"away_team_all_score/count_season",
                                                                     "Count_J1_Season":"away_team_Count_J1_Season","diff_match_date":"away_team_diff_match_date"})
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_num_match"] = home_df2["home_team_season_num_match"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_season_num_match"] = away_df2["away_team_season_num_match"] 
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_point"] = home_df2["home_team_season_point"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_season_point"] = away_df2["away_team_season_point"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_get_score"] = home_df2["home_team_season_get_score"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_season_get_score"] = away_df2["away_team_season_get_score"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_season_lost_score"] = home_df2["home_team_season_lost_score"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_season_lost_score"] = away_df2["away_team_season_lost_score"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_point_sum"] = home_df2["home_team_Curren4match_point_sum"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_point_sum"] = away_df2["away_team_Curren4match_point_sum"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_point_mean"] = home_df2["home_team_Curren4match_point_mean"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_point_mean"] = away_df2["away_team_Curren4match_point_mean"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_point_std"] = home_df2["home_team_Curren4match_point_std"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_point_std"] = away_df2["away_team_Curren4match_point_std"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_get_score_sum"] = home_df2["home_team_Curren4match_get_score_sum"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_get_score_sum"] = away_df2["away_team_Curren4match_get_score_sum"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_get_score_mean"] = home_df2["home_team_Curren4match_get_score_mean"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_get_score_mean"] = away_df2["away_team_Curren4match_get_score_mean"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_get_score_std"] = home_df2["home_team_Curren4match_get_score_std"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_get_score_std"] = away_df2["away_team_Curren4match_get_score_std"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_lost_score_sum"] = home_df2["home_team_Curren4match_lost_score_sum"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_lost_score_sum"] = away_df2["away_team_Curren4match_lost_score_sum"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_lost_score_mean"] = home_df2["home_team_Curren4match_lost_score_mean"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_lost_score_mean"] = away_df2["away_team_Curren4match_lost_score_mean"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Curren4match_lost_score_std"] = home_df2["home_team_Curren4match_lost_score_std"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Curren4match_lost_score_std"] = away_df2["away_team_Curren4match_lost_score_std"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_before_match_point"] = home_df2["home_team_before_match_point"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_before_match_point"] = away_df2["away_team_before_match_point"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_before_match_get_score"] = home_df2["home_team_before_match_get_score"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_before_match_get_score"] = away_df2["away_team_before_match_get_score"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_before_match_lost_score"] = home_df2["home_team_before_match_lost_score"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_before_match_lost_score"] = away_df2["away_team_before_match_lost_score"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_all_point"] = home_df2["home_team_all_point"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_all_point"] = away_df2["away_team_all_point"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_all_get_score_sum"] = home_df2["home_team_all_get_score_sum"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_all_get_score_sum"] = away_df2["away_team_all_get_score_sum"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_all_get_score_mean"] = home_df2["home_team_all_get_score_mean"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_all_get_score_mean"] = away_df2["away_team_all_get_score_mean"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_all_point/count_season"] = home_df2["home_team_all_point/count_season"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_all_point/count_season"] = away_df2["away_team_all_point/count_season"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_all_score/count_season"] = home_df2["home_team_all_score/count_season"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_all_score/count_season"] = away_df2["away_team_all_score/count_season"]
    #match_info_df.loc[match_info_df["home_team"]==team,"home_team_Count_J1_Season"] = home_df2["home_team_Count_J1_Season"]
    #match_info_df.loc[match_info_df["away_team"]==team,"away_team_Count_J1_Season"] = away_df2["away_team_Count_J1_Season"]
    match_info_df.loc[match_info_df["home_team"]==team,"home_team_diff_match_date"] = home_df2["home_team_diff_match_date"]
    match_info_df.loc[match_info_df["away_team"]==team,"away_team_diff_match_date"] = away_df2["away_team_diff_match_date"]
    del home_df,away_df,team_df,home_df2,away_df2
match_info_df
```

```python
venue_ful=["日産スタジアム","埼玉スタジアム2002","味の素スタジアム","万博記念競技場","ニッパツ三ツ沢球技場","京都市西京極総合運動公園陸上競技場兼球技場","広島ビッグアーチ","山梨中銀スタジアム","県立カシマサッカースタジアム","アウトソーシングスタジアム日本平","ユアテックスタジアム仙台","フクダ電子アリーナ","等々力陸上競技場","NACK5スタジアム大宮","ベストアメニティスタジアム","日立柏サッカー場","ヤマハスタジアム(磐田)","名古屋市瑞穂陸上競技場"]
def get_rain(weather):
    if weather[0] in ["雨", "雪", "雷"]:
        return 1
        
    else:
        return 0

def get_sunny(weather):
    if weather=="晴":
        return 1

    else:
        return 0


def get_weather(df):
    df["have_rain"] = df["weather"].apply(get_rain)
    df["have_sunny"] = df["weather"].apply(get_sunny)

    return df

train = get_weather(train)
test = get_weather(test)
```

```python
def lgb_training(x_train: pd.DataFrame, y_train: pd.DataFrame, x_valid: pd.DataFrame, y_valid: pd.DataFrame,
                 features: list, categorical_features: list):
    clf = lgbm.LGBMRegressor(**CFG.lgb_params)
    clf.fit(x_train, y_train,
                eval_set=[(x_valid, y_valid)],
                eval_metric=CFG.metric,
                verbose = CFG.verbose,
              #early_stopping_rounds=100,
              #categorical_feature = categorical_features
              )
    # Predict validation
    valid_pred = clf.predict(x_valid)
    return clf, valid_pred

def xgb_training(x_train: pd.DataFrame, y_train: pd.DataFrame, x_valid: pd.DataFrame, y_valid: pd.DataFrame,
                 features: list, categorical_features: list):
    clf = xgb.XGBRegressor(**CFG.xgb_params)
    clf.fit(x_train, y_train,
                eval_set=[(x_valid, y_valid)],
                eval_metric=CFG.metric,
                verbose = CFG.verbose,
               early_stopping_rounds=CFG.early_stopping_rounds,
              #categorical_feature = categorical_features
              )
    # Predict validation
    valid_pred = clf.predict(x_valid)
    return clf, valid_pred

def ctb_training(x_train: pd.DataFrame, y_train: pd.DataFrame, x_valid: pd.DataFrame, y_valid: pd.DataFrame,
                 features: list, categorical_features: list):
    clf = CatBoostRegressor(**CFG.ctb_params)
    clf.fit(x_train, y_train,
          eval_set=[(x_train,y_train),(x_valid, y_valid)],
          early_stopping_rounds = CFG.early_stopping_rounds,
          use_best_model=True
          )
    # Predict validation
    valid_pred = clf.predict(x_valid)
    return clf, valid_pred

def tabnet_training(x_train: pd.DataFrame, y_train: pd.DataFrame, x_valid: pd.DataFrame, y_valid: pd.DataFrame,
                 features: list, categorical_features: list):
    clf = TabNetRegressor(verbose=CFG.verbose,seed=CFG.seed)
    y_train = np.log1p(y_train)
    y_valid = np.log1p(y_valid)
    clf.fit(X_train=x_train.to_numpy(), y_train=y_train.to_numpy().reshape(-1, 1),
              eval_set=[(x_valid.to_numpy(), y_valid.to_numpy().reshape(-1, 1))],
              patience=300, max_epochs=CFG.max_epochs,
              eval_metric=[CFG.metric])
    valid_pred = clf.predict(x_valid.to_numpy())
    valid_pred = np.expm1(valid_pred)
    return clf, valid_pred
    
    
def cv_training(method: str, train_df: pd.DataFrame, features: list, categorical_features: list):
    scores, preds = [],[]
    oof = pd.DataFrame()
    if method == 'lightgbm':
        lgb_importances = pd.DataFrame()
    kf = KFold(n_splits=CFG.n_folds ,shuffle=True, random_state=CFG.seed)
    for fold ,(tr_idx,va_idx) in enumerate(kf.split(train_df)):
        print("="*20,f" {method} training fold {fold + 1} ","="*20)
        
        train_folds = train_df.iloc[tr_idx]
        valid_folds = train_df.iloc[va_idx]
        x_train = train_folds[features]
        x_valid = valid_folds[features]
        y_train = train_folds[CFG.target_col]
        y_valid = valid_folds[CFG.target_col]
            
        if method == 'lightgbm':
            model, valid_pred = lgb_training(x_train, y_train, x_valid, y_valid, features, categorical_features)
            fi_tmp = pd.DataFrame()
            fi_tmp['feature'] = x_train.columns
            fi_tmp['importance'] = model.feature_importances_
            fi_tmp['fold'] = fold
            #変更点
            lgb_importances = []
            lgb_importances.append(fi_tmp)
            lgb_importances=pd.concat(lgb_importances)
        if method == 'xgboost':
            model, valid_pred = xgb_training(x_train, y_train, x_valid, y_valid, features, categorical_features)
        if method == 'catboost':
            model, valid_pred = ctb_training(x_train, y_train, x_valid, y_valid, features, categorical_features)
        if method == 'tabnet':
            model, valid_pred = tabnet_training(x_train, y_train, x_valid, y_valid, features, categorical_features)

        # Save best model
        pickle.dump(model, open(os.path.join(CFG.MODEL_DATA_PATH, f'{method}_fold{fold + 1}_seed{CFG.seed}_ver{CFG.VER}.pkl'), 'wb'))
            

        score = np.sqrt(mean_squared_error(y_valid,valid_pred))
        print(f"RMSE : {score}")
        scores.append(score)
            
        valid_folds[f"{method}_pred"] = valid_pred
        
        oof = pd.concat([oof,valid_folds])
        
        del x_train,y_train,x_valid,y_valid
        
    print(f"Mean Score {np.mean(scores)}")
    
    oof.to_csv(os.path.join(CFG.OOF_DATA_PATH, f'oof_{method}_seed{CFG.seed}_ver{CFG.VER}.csv'), index = False)
    
    score = np.sqrt(mean_squared_error(oof["attendance"],oof[f"{method}_pred"]))
    print(f"RMSE : {score}")
    
    if method == 'lightgbm':
        order = list(lgb_importances.groupby('feature').mean().sort_values('importance', ascending=False).index)
        fig = plt.figure(figsize=(16, 32), tight_layout=True)
        sns.barplot(x="importance", y="feature", data=lgb_importances.groupby('feature').mean().reset_index(), order=order)
        plt.title("LightGBM feature importances")
        
```
